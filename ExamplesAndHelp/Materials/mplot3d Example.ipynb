# mplot3d tutorial

### Import libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
from sklearn.decomposition import PCA
%matplotlib notebook
%matplotlib notebook

```

### Create a simple plot

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d') # figure 1 by 1 by 1
```

### Spawn some data

```python
X = np.random.random((10, 3))
X
```

### Plot the data

PCA will convert np.array into acceptable data object

```python
pca = PCA(n_components=3)
```

Load the data

```python
pca.fit(X)
```

Transform the data

```python
X_pca = pca.transform(X)
```

Plot the data

```python
ax.scatter3D(X_pca[:, 0], X_pca[:, 1], X_pca[:,2])
```

```python
fig.show()
```

### The output will appear in the upper plot
