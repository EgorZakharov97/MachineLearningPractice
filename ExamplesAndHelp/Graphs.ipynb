```python
import matplotlib.pyplot as plt
%matplotlib notebook
%matplotlib notebook

x = np.linspace(-4, 4)
y1, y2 = x**2, x**3

fig, ax = plt.subplots()
ax.plot(x, y1, 'red')
ax.plot(x, y2, 'blue')
ax.set_title('Quadratic and cubic powers')
ax.set_xlabel('x values')
ax.set_ylabel('y values')
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)

axes[0,0].set_title('Upper Left')
axes[0,1].set_title('Upper Right')
axes[1,0].set_title('Lower Left')
axes[1,1].set_title('Lower Right')

# set marks to empry
for ax in axes.flat:
    ax.set(xticks=[], yticks=[])
```

```python
fig, axes = plt.subplots(nrows=2, ncols=5, figsize=(14, 4.5))
for i, ax in enumerate(axes.flatten()):
    ax.set_title("x to the {:d}".format(i+1))
    ax.plot(x, x**(i+1))
    
fig.tight_layout()
```

```python
# Visualization with pandas
%matplotlib inline
table = pd.read_csv('CC_GENERAL.csv', index_col='CUST_ID')
table['PURCHASES_FREQUENCY'].value_counts().hist(edgecolor='black', bins=10).set_title('Histogram')
```

```python
table['BALANCE_FREQUENCY'].value_counts().plot(kind='bar')
```

```python
table.plot.scatter(x='BALANCE', y='INSTALLMENTS_PURCHASES')
```

```python
# How to modify figure more with more flexibility
fig, ax = plt.subplots()
table.plot.scatter(x='BALANCE', y='INSTALLMENTS_PURCHASES', ax=ax)
ax.set_title("Title")
ax.set_xlabel("X")
ax.set_ylabel("Y")
```

##Different types of plots in panda
* 'bar' or 'barh' for bar plots
* 'hist' for histogram
* 'box' for boxplot
* 'kde' or 'density' for density plots
* 'area' for area plots
* 'scatter' for scatter plots
* 'hexbin' for hexagonal plots
* 'pie' for pie plots
